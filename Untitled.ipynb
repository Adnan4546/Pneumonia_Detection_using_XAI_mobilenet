```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.layers import Input, Flatten, Dropout, Dense, BatchNormalization
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras. optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
from PIL import Image, ImageEnhance
import cv2
```

```python
# Define paths based on your folder structure
train_path = 'train'
val_path = 'val'
test_path = 'test'

# Augmentation for Training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Only Rescaling for Val/Test
val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path, target_size=(224, 224), batch_size=32, class_mode='binary')

val_generator = val_test_datagen.flow_from_directory(
    val_path, target_size=(224, 224), batch_size=32, class_mode='binary')

test_generator = val_test_datagen.flow_from_directory(
    test_path, target_size=(224, 224), batch_size=32, class_mode='binary', shuffle=False)
```

```python
# MobileNetV2 expects 224x224 images by default
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze the weights

model = Sequential([
    base_model,
    # MobileNetV2 output is already quite optimized, so we use GlobalAveragePooling
    tf.keras.layers.GlobalAveragePooling2D(), 
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.0001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# 1. Define your callbacks FIRST
checkpoint = ModelCheckpoint("my_model.keras", monitor='val_accuracy', save_best_only=True, mode='max')
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6, verbose=1)

# Group them into a list
callbacks_list = [checkpoint, early_stop, reduce_lr]

# 2. Unfreeze the base model
base_model.trainable = True

# 3. Re-freeze the bottom layers, leaving only the top layers trainable
# MobileNetV2 has 155 layers. Let's unfreeze the top 30.
for layer in base_model.layers[:-30]:
    layer.trainable = False

# 4. Recompile
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)

# 5. Continue training using the defined callbacks_list
history_finetune = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,
    callbacks=callbacks_list # <-- Use the list we created above
)
```

```python
# Predictions
predictions = model.predict(test_generator)
y_pred = (predictions > 0.5).astype(int)
y_true = test_generator.classes

# 1. Classification Report
print(classification_report(y_true, y_pred, target_names=test_generator.class_indices.keys()))

# 2. Confusion Matrix Heatmap
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Normal', 'Pneumonia'], 
            yticklabels=['Normal', 'Pneumonia'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name="out_relu"):
    # 1. Access the base model (MobileNetV2) inside the Sequential wrapper
    base_model = model.layers[0]
    
    # 2. Construct the Grad-CAM model using the MAIN model's input
    # This connects the whole pipeline from Input -> MobileNet -> Dense layers
    grad_model = tf.keras.models.Model(
        inputs=[model.input],
        outputs=[base_model.get_layer(last_conv_layer_name).output, model.output]
    )

    # 3. Record gradients
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        # Target the prediction score
        class_channel = preds[:, 0]

    # 4. Compute gradients and generate heatmap
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize heatmap
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-10)
    return heatmap.numpy()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def save_and_display_gradcam(img_path, heatmap, alpha=0.4):
    # 1. Load the original image
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224, 224))

    # 2. Resize the 7x7 heatmap to match the 224x224 image size
    heatmap = cv2.resize(heatmap, (img.shape[1], img.shape[0]))

    # 3. Rescale heatmap to a range 0-255
    heatmap = np.uint8(255 * heatmap)

    # 4. Use jet colormap to colorize heatmap
    jet = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    # 5. Superimpose the heatmap on original image
    superimposed_img = jet * alpha + img
    superimposed_img = np.clip(superimposed_img, 0, 255).astype(np.uint8)

    # 6. Display the result
    plt.imshow(cv2.cvtColor(superimposed_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("AI Focus Area ")
    plt.show()
```

```python
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array

def predict_pneumonia(img_path, trained_model, threshold=0.5):
    """
    Takes an image path and a trained model, and predicts if the X-ray shows pneumonia.
    """
    try:
        # 1. Load the image and resize to MobileNetV2's required input size
        img = load_img(img_path, target_size=(224, 224))
        
        # 2. Convert image to a Numpy array and normalize (rescaling 1./255)
        img_array = img_to_array(img) / 255.0
        
        # 3. Add the batch dimension (Keras expects [batch_size, height, width, channels])
        img_array = np.expand_dims(img_array, axis=0)
        
        # 4. Make the prediction (verbose=0 hides the progress bar for cleaner output)
        prediction_prob = trained_model.predict(img_array, verbose=0)[0][0]
        
        # 5. Interpret the probability
        # Since we mapped class_mode='binary', typically 0 = Normal, 1 = Pneumonia
        if prediction_prob >= threshold:
            diagnosis = "PNEUMONIA"
            confidence = prediction_prob * 100
        else:
            diagnosis = "NORMAL"
            confidence = (1.0 - prediction_prob) * 100
            
        return diagnosis, confidence
        
    except Exception as e:
        return f"Error processing image: {e}", 0.0
```
